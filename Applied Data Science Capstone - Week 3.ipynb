### Import pandas library. Read the wikipedia page using pd.read_html and assign it to tables

```python
import pandas as  pd

tables=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
tables
```

### We require only the first table. Hence assign the first table to df and display df.

```python
df=tables[0]
df
```

### Set index as Borough using set_index

```python
df.set_index('Borough',inplace=True)
df
```

### Drop the values 'Not assigned' in Borough

```python
df=df.drop('Not assigned')
```

```python
df
```

### Reset index to get back the index values

```python
df=df.reset_index()
```

```python
df
```

### Use the reindex method to swap columns Borough and Postal Code

```python
columns_titles = ["Postal Code","Borough",'Neighborhood']
df=df.reindex(columns=columns_titles)
```

```python
df
```

### Check for 'Not assigned' values in the column 'Neighborhood'

```python
c=0
for i in df['Neighborhood']:
    if('Not assigned' in i):
        c=1
        
if c==0:
    print('All neighborhoods are assigned')
    
    
```

### Extract data whose borough contains 'Toronto'

```python

toronto_data=df[df['Borough'].str.contains('Toronto')]
toronto_data=toronto_data.reset_index(drop=True)
```

```python
toronto_data.head()
```

### Visualise the toronto map

```python
import folium
toronto_map=folium.Map(location=[43.6532,-79.3832],zoom_start=12)
colors = ["Red", "Blue", "Green", "Purple"]
boroughs = set(toronto_data['Borough'])
count = 0
for b in boroughs:
    for lat,lng,label,bor in zip(toronto_data['Latitude'],toronto_data['Longitude'],toronto_data['Neighborhood'], toronto_data['Borough']):
        if (bor == b):
            label=folium.Popup(label,parse_html=False)
            folium.CircleMarker([lat,lng],radius=7,popup=label,color=colors[count],
                fill=True,
                fill_color=colors[count],
                fill_opacity=0.7,
                parse_html=False).add_to(toronto_map)
    count = count + 1
toronto_map
```
