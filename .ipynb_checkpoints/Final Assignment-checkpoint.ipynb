```python
import pandas as pd
```

```python
df_2011=df[['Name','2011']]
```

```python
df_2011
```

```python
for i, name in enumerate(df_2011['Name']):
    si = name.find('(')
    if (si != -1):
        si = si - 1
        df_2011['Name'][i] = name[:si]
        print(df_2011['Name'][i])
```

```python
df_2011=df_2011.reset_index()

```

```python
df_2011
```

```python
latitude=[]
longitude=[]
geolocator = Nominatim(user_agent="foursquare_agent")
for i,name in enumerate(df_2011['Name']):
    address=name
    location = geolocator.geocode(address)
    latitude.extend([location.latitude]) 
    longitude.extend([location.longitude])
    
longitude
```

```python
dict={'Latitude':latitude,'Longitude':longitude}
dict
latlng_data=pd.DataFrame(dict)
df_2011=df_2011.join(latlng_data)
df_2011
```

```python
address='Tamil Nadu'
location = geolocator.geocode(address)
latitude=location.latitude
longitude=location.longitude
print(latitude,longitude)
```

```python
c=0
import folium
tn_map=folium.Map(location=[latitude,longitude],zoom_start=7)
for lat,lng,label in zip(df_2011['Latitude'],df_2011['Longitude'],df_2011['Name']):
    label=folium.Popup(label,parse_html=False)
    folium.CircleMarker([lat,lng],radius=7,popup=label,color='blue',
    fill=True,
    fill_color='blue',
    fill_opacity=0.7,
    parse_html=False).add_to(tn_map)
    c=c+1
print(c)
tn_map
```

```python
dataframe.shape
```

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

dataframe['categories'] = dataframe.apply(get_category_type, axis=1)
dataframe
```
